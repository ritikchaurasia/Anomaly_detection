## Create the Model

```python
# Remove 'timestamp' column for model training
columns_to_use = [
    'cpu_utilization_24ae8d', 'cpu_utilization_53ea38', 'cpu_utilization_5f5533',
    'cpu_utilization_77c1ca', 'cpu_utilization_825cc2', 'cpu_utilization_ac20cd',
    'cpu_utilization_c6585a', 'cpu_utilization_fe7f93', 'disk_write_bytes_1ef3de',
    'disk_write_bytes_c0d644', 'ec2_network_in_257a54', 'ec2_network_in_5abac7',
    'elb_request_count_8c0756', 'grok_asg_anomaly', 'iio_us-east-1_i-a2eb1cd9_NetworkIn',
    'rds_cpu_utilization_cc0c53', 'rds_cpu_utilization_e47b3b'
]

# Prepare the data (excluding 'timestamp')
X = merged_df[columns_to_use]

# Fit the IsolationForest model
model = IsolationForest(contamination=0.01, random_state=42)
merged_df['anomaly_pred'] = model.fit_predict(X)

# Convert model predictions: -1 (anomaly) becomes 1, and 1 (normal) becomes 0
merged_df['anomaly_pred'] = merged_df['anomaly_pred'].apply(lambda x: 1 if x == -1 else 0)

# Dummy ground truth 'anomaly' column for testing (replace with actual data if available)
if 'anomaly' not in merged_df.columns:
    merged_df['anomaly'] = np.random.choice([0, 1], size=len(merged_df))  # Replace with true labels

# Assuming 'anomaly' column contains the ground truth values
y_true = merged_df['anomaly']  # Ground truth
y_pred = merged_df['anomaly_pred']  # Predicted values

# Calculate precision, recall, and F1 score
precision = precision_score(y_true, y_pred)
recall = recall_score(y_true, y_pred)
f1 = f1_score(y_true, y_pred)

# Output results
print(f"Number of anomalies detected: {y_pred.sum()}")
print(f"Precision: {precision}")
print(f"Recall: {recall}")
print(f"F1 Score: {f1}")
```

### Using some parameter of the Algorithm to improve the Performances of the model

```python
model = IsolationForest(
    contamination=0.01,  # Try different contamination levels
    n_estimators=200,    # Increase number of estimators
    max_samples=0.9,     # Sample 90% of the data for each tree
    random_state=42
)
merged_df['anomaly_pred'] = model.fit_predict(X)

# Convert model predictions: -1 (anomaly) becomes 1, and 1 (normal) becomes 0
merged_df['anomaly_pred'] = merged_df['anomaly_pred'].apply(lambda x: 1 if x == -1 else 0)

# Dummy ground truth 'anomaly' column for testing (replace with actual data if available)
if 'anomaly' not in merged_df.columns:
    merged_df['anomaly'] = np.random.choice([0, 1], size=len(merged_df))  # Replace with true labels

# Assuming 'anomaly' column contains the ground truth values
y_true = merged_df['anomaly']  # Ground truth
y_pred = merged_df['anomaly_pred']  # Predicted values

# Calculate precision, recall, and F1 score
precision = precision_score(y_true, y_pred)
recall = recall_score(y_true, y_pred)
f1 = f1_score(y_true, y_pred)

# Output results
print(f"Number of anomalies detected: {y_pred.sum()}")
print(f"Precision: {precision}")
print(f"Recall: {recall}")
print(f"F1 Score: {f1}")
```

```python
# Plot multiple metrics in one plot
def plot_multiple_metrics(df, columns, anomalies_column='anomaly'):
    plt.figure(figsize=(14, 8))

    for column in columns:
        plt.plot(df.index, df[column], label=column)

    # Highlight anomalies
    anomalies = df[df[anomalies_column] == 1]
    for column in columns:
        plt.scatter(anomalies.index, anomalies[column], label=f'{column} Anomalies', marker='x')

    plt.title("Multiple Metrics with Anomalies")
    plt.xlabel('Timestamp')
    plt.ylabel('Value')
    plt.legend()
    plt.show()

# Plot example for multiple columns
metrics = ['cpu_utilization_24ae8d', 'disk_write_bytes_1ef3de', 'ec2_network_in_257a54']
plot_multiple_metrics(merged_df, metrics)
```

```python
# Function to plot time series data with anomalies highlighted
def plot_anomalies(df, column, anomalies_column='anomaly_pred'):
    plt.figure(figsize=(12, 6))
    plt.plot(df.index, df[column], label='Data', color='blue')

    # Highlight anomalies in red
    anomalies = df[df[anomalies_column] == 1]
    plt.scatter(anomalies.index, anomalies[column], color='red', label='Anomalies', marker='x')

    plt.title(f"{column} with Anomalies Highlighted")
    plt.xlabel('Timestamp')
    plt.ylabel(column)
    plt.legend()
    plt.show()

# Plot example for 'cpu_utilization_24ae8d' column
plot_anomalies(merged_df, 'cpu_utilization_24ae8d')
```
